```python
url = 'https://burger-king-kupon.ru/calories/'
info = r.get(url)
soup = BeautifulSoup(info.text, 'html.parser')
# soup
```

```python
count = 1
food = []
link = []
b = []
jir = []
c = []
cal = []
for i in range(len(soup.findAll('table'))):
  for j in soup.findAll('table')[i].findAll('td'):
    if count == 1:
      food.append(j.contents[0].contents[0])
      for_img = r.get(j.contents[0].get('href')).text
      try:      
        link.append(re.findall('(?s)<figure class="image main-picture p-4">\r\n\t\t\t\t\t\t\t  \t<img width="256" height="256" src="(.*)" class="product-image', for_img)[0])
      except:
        link.append(re.findall('(?s)<figure class="image main-picture p-4">\r\n\t\t\t\t\t\t\t  \t<img width="256" height="165" src="(.*)" class="product-image', for_img)[0])      
      count += 1
    elif count == 2:
      b.append(j.contents[0])
      count += 1
    elif count == 3:
      jir.append(j.contents[0])
      count += 1
    elif count == 4:
      c.append(j.contents[0])
      count += 1
    elif count == 5:
      cal.append(j.contents[0])
      count = 1

final_table = pd.DataFrame({'Еда': food, 'Ссылка': link, 'Белки': b, 'Жиры': jir, 'Углеводы': c, 'Калории': cal})
final_table['Углеводы'] = final_table['Углеводы'].replace('г.', '', regex = True)

name = []
for i in range(len(final_table['Еда'])):
  name.append('_'.join(final_table['Еда'][i].split(' ')).replace('/', ''))
final_table['name'] = name
final_table
```
